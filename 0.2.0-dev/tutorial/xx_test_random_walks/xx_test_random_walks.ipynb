# Higher Order Random Walk

```python
g3 = pp.Graph.from_edge_list([
    ['a','b'],
    ['b','c'],
    ['c','a'],
    ['c','d'],
    ['d','a']
    ])

g3.data['edge_weight'] = torch.tensor([[1],[1],[2],[1],[1]])
```

```python
pp.plot(g3, node_label= [g3.mapping.to_id(x) for x in range(g3.n)]);
```

```python
g3.mapping.id_to_idx
```

```python
paths = pp.DAGData(g3.mapping)
paths.append_walk(['a','b','c'],weight=1)
paths.append_walk(['b','c','a'],weight=1)
paths.append_walk(['b','c','d'],weight=0.2)
paths.append_walk(['c','a','b'],weight=1)
paths.append_walk(['c','d','a'],weight=0.2)
paths.append_walk(['d','a','b'],weight=1)
```

```python
paths.dags
```

```python
m = pp.MultiOrderModel.from_DAGs(paths, max_order=3)
g3_ho = m.layers[2]
```

```python
pp.plot(g3_ho, node_label = [g3_ho.mapping.to_id(x) for x in range(g3_ho.n)]);
```

```python
rw = HigherOrderRandomWalk(g3_ho, g3, weight=True)
```

```python
data = rw.run_experiment(steps=100, runs=list(g3_ho.nodes))
```
